```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open('MoNuSeg 2018 Training Data\Tissue Images\TCGA-18-5592-01Z-00-DX1.tif')
```

```python
img
```

```python
img_array = np.array(img)
```

```python
x_min = 250
x_max = 500
y_min = 250
y_max = 500

patch = img_array[x_min: x_max, y_min: y_max]
temp_img = img_array.copy()

temp_img[x_max - 1: x_max + 1, y_min: y_max] = [0, 0, 0]
temp_img[x_min: x_max, y_max - 1: y_max + 1] = [0, 0, 0]
temp_img[x_min - 1: x_min + 1, y_min: y_max] = [0, 0, 0]
temp_img[x_min: x_max, y_min - 1: y_min + 1] = [0, 0, 0]


plt.figure(figsize = (20, 10))
ax1 = plt.subplot(1, 2, 1)
plt.imshow(temp_img)
ax2 = plt.subplot(1, 2, 2)
plt.imshow(patch)
plt.show()
```
